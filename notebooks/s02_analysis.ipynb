# Setup

```python
import os, sys, configparser


settings_parser = configparser.ConfigParser()
settings_parser.read("localsettings.ini")
MAIN_DATA_FOLDER = settings_parser.get("global", "data_path")
MIENC_PATH = settings_parser.get("global", "mienc_path")
cache_dir = os.path.join(MAIN_DATA_FOLDER, "cache")
if not os.path.isdir(cache_dir):
    os.makedirs(cache_dir)
config_ini = os.path.join(MAIN_DATA_FOLDER, "config.ini")

sys.path.append(MIENC_PATH)
from mienc import NonLinearEstimator as NLE
```

# fMRI

```python
regions = [
    10,
    30,
    50,
    70,
    100,
    150,
    200,
    230,
    270,
    300,
    350,
    400,
    450,
    500,
    550,
    600,
    650,
    700,
    750,
    800,
    850,
    900,
    950,
]
nle = NLE(
    config_file=config_ini,
    bins=0,
    surrogates=99,
    cache=cache_dir,
    save_out=20,
    dataset="clean_cra",
    verbose=True,
)
for region in regions:
    nle.estimate(
        display=False, dataset_sub=str(region), extended_stats=True, compute_shadow=True
    )
```

# fMRI test-retest

```python
nle = NLE(
    config_file=config_ini,
    bins=0,
    surrogates=99,
    cache=cache_dir,
    save_out=True,
    dataset="lemon_aal",
    verbose=True,
)
for session in range(3):
    nle.estimate(
        display=False,
        dataset_sub=str(session),
        extended_stats=True,
        compute_shadow=True,
    )
```
