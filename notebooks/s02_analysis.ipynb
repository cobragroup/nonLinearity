# Setup

```python
import os, sys, configparser, tqdm
import numpy as np

settings_parser = configparser.ConfigParser()
settings_parser.read("localsettings.ini")
RANDOM_SEED = settings_parser.getint("global", "random_seed")
MAIN_DATA_FOLDER = settings_parser.get("global", "data_path")
MIENC_PATH = settings_parser.get("global", "mienc_path")
cache_dir = os.path.join(MAIN_DATA_FOLDER, "cache")
if not os.path.isdir(cache_dir):
    os.makedirs(cache_dir)
config_ini = os.path.join(MAIN_DATA_FOLDER, "config.ini")

sys.path.append(MIENC_PATH)
from mienc import NonLinearEstimator as NLE
```

# fMRI

```python
GOOD_SESSIONS = [
    831882777,
    816200189,
    771160300,
    786091066,
    779839471,
    778998620,
    781842082,
    778240327,
    793224716,
    839068429,
    794812542,
    768515987,
    767871931,
    840012044,
    766640955,
    847657808,
]

TIMES = [125 * 2**i for i in range(7)]
splits = [2, 3, 5, 6, 10]
base_length = 1800000

total = len(GOOD_SESSIONS) * 3 * np.sum(splits)
with tqdm.tqdm(total=total) as pbar:
    for session in GOOD_SESSIONS:
        pbar.set_description(f"Non-stationarity, session {session}")
        nle = NLE(
            config_file=config_ini,
            surrogates=99,
            jitter=True,
            cache=cache_dir,
            save_out=True,
            dataset=f"spiking_{session}",
            verbose=False,
            random_state=RANDOM_SEED,
        )
        for t in TIMES[:3]:
            total_length = base_length // t
            for split in splits:
                part_length = total_length // split
                for part in range(split):
                    nle.estimate(
                        bins=0,
                        truncate_input=[part_length * part, part_length * (part + 1)],
                        suffix=f"spl{split:02}_part{part:02}",
                        display=False,
                        dataset_sub=f"{t:04}",
                        extended_stats=True,
                        compute_shadow=True,
                    )
                    pbar.update()
```
