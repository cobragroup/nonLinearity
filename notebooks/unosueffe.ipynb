```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
T = 200000000 # 2000 s
fs = 200000
time_series_1 = np.zeros(T)
current_time = 0

tau1 = 5000
tau2 = 3000

while current_time<T:
    current_time += int(round(-tau1*np.log(np.random.rand()),0))
    t = int(round(-tau2*np.log(np.random.rand()),0))
    time_series_1[current_time:current_time+t] = 1
    current_time += t
```

```python
fs = 200000
f, Pxx_spec = signal.welch(time_series_1, fs, 'flattop', 1024, scaling='spectrum')
```

```python

fig, ax = plt.subplots(1, 2, figsize=(9,5), gridspec_kw={"width_ratios":(2,1)})
plt.sca(ax[0])
plt.loglog(f, np.sqrt(Pxx_spec))
plt.plot([3000, 30000], [.1,.01], "--k", label=r"$P=A\cdot\frac{1}{f}$")
plt.xlabel('frequency [Hz]')
plt.ylabel('Linear spectrum [V RMS]')
plt.title('Power spectrum (scipy.signal.welch)')
plt.legend()
plt.sca(ax[1])
plt.plot(np.arange(100000)/fs, time_series_1[:100000])
plt.xlabel("t [s]")
plt.ylabel("State")
plt.title("Sample sequence")
plt.suptitle("Single bistable system "+r"$\tau_0=25 ms, \tau_1=15 ms$")
plt.show()
np.log10(.1/.01)/np.log10(100/10)
```

```python
T = 2000000 # 1000 s
fs = 2000
final_ts = np.zeros(T)

tau1 = 25
tau2 = 15

for i in range(1000):
    current_time = 0
    time_series = np.zeros(T)
    while current_time<T:
        current_time += int(round(-tau1*np.log(np.random.rand()),0))
        t = int(round(-tau2*np.log(np.random.rand()),0))
        time_series[current_time:current_time+t] = 1
        current_time += t
    final_ts += time_series
```

```python
f, Pxx_spec = signal.welch(final_ts, fs, 'flattop', 1024, scaling='spectrum')
fig, ax = plt.subplots(1, 2, figsize=(9,5), gridspec_kw={"width_ratios":(2,1)})
plt.sca(ax[0])
plt.loglog(f, np.sqrt(Pxx_spec))
plt.plot([31.4, 314], [3.14,.314], "--k", label=r"$P=A\cdot\frac{1}{f}$")
plt.xlabel('frequency [Hz]')
plt.ylabel('Linear spectrum [V RMS]')
plt.title('Power spectrum (scipy.signal.welch)')
plt.legend()
plt.sca(ax[1])
plt.plot(np.arange(1000)/2000+0.5, final_ts[1000:2000])
plt.xlabel("t [s]")
plt.ylabel("State")
plt.title("Sample sequence")
plt.suptitle("1000 bistable systems "+r"$\tau_0^i=50 ms, \tau_1^i=30 ms \:\forall i$")
plt.show()
np.log10(.1/.01)/np.log10(100/10)
```

```python
a = 2
wl = []
sd = []
while a<T/10:
    twl = int(a)
    nw = min(1000, int(T/twl))
    spa = int(T/nw)
    wl.append(twl)
    tsd = []
    for i in range(nw):
        tsd.append(np.std(final_ts[i*spa:i*spa+twl]))
    sd.append(np.mean(tsd))
    a+=1.5
```
